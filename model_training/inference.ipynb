```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
path_to_model = "best.pt"
path_to_inf_image = "000019.png"
```

```python
model = YOLO(path_to_model)
results = model(path_to_inf_image)
```

```python
for result in results:
    plot = result.plot()

    plt.imshow(plot)
    plt.axis("off")
    plt.show()
```
